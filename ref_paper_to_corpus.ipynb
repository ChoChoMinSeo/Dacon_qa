```python
import fitz
import glob
import re
from hanspell import spell_checker
from tqdm import tqdm
```

```python
paths = glob.glob('./reference_papers/*.pdf')
```

```python
pages = []
with open('./reference_papers/context.txt','r') as t:
    for line in t.readlines():
        pages.append(list(map(int,line.split(' '))))
```

```python
# page by page
def merge_normal_pdf(page_texts):
    temp = ''
    for sentence in page_texts[1:]:
        if len(re.sub(r'\d', '', sentence))<4:
            pass
        else:
            temp+=sentence
    return temp
def merge_not_normal_pdf(page_texts):
    temp = ''
    for sentence in page_texts[1:]:
        temp+=sentence    
    return temp
```

```python
pdf_idx=1
print(paths[pdf_idx])
res=''
tot = 0
with fitz.open(paths[pdf_idx]) as doc:
    for i in range(pages[pdf_idx][0]-1,pages[pdf_idx][1]):#pages[pdf_idx][0]-1,pages[pdf_idx][1]
        text_in_page = doc[i].get_text()
        tot+=len(text_in_page)
        # print(text_in_page.split('\n'))
        res+=merge_not_normal_pdf(text_in_page.split('\n'))
print(res)
print(len(res),tot)
```

```python
not_normal=[1,2,4,9]
fin=[0,1,2, 3,4,5,7,8]## 6번은 영어가 많아서 안되는듯 하다.
for pdf_idx in tqdm(range(len(paths))):
    if pdf_idx in fin:
        continue
    print(pdf_idx, paths[pdf_idx]+'\n')
    tmp=''
    res=''
    tot = 0
    with fitz.open(paths[pdf_idx]) as doc:
        for i in range(pages[pdf_idx][0]-1,pages[pdf_idx][1]):#pages[pdf_idx][0]-1,pages[pdf_idx][1]
            text_in_page = doc[i].get_text()
            tot+=len(text_in_page)
            if pdf_idx not in not_normal:
                tmp += merge_normal_pdf(text_in_page.split('\n'))
            else:
                tmp +=merge_not_normal_pdf(text_in_page.split('\n'))
        print(tmp)
        for i in range(len(tmp)//100+1):
            res += spell_checker.check(tmp[i*100:min((i+1)*100,len(tmp))]).only_checked()
    print(len(tmp),len(res),tot)
    print(res)
    print('\n')
    with open(f'./reference_papers/{pdf_idx+1} corpus.txt','w',encoding='utf-8') as f:
        f.write(res)
```
